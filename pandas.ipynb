```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
}

df = pd.DataFrame(data)

df['Gender'] = ['F', 'M']

print(df)

df.drop('Age', axis=1, inplace=True)

print(df)
```

```python
import pandas as pd

data = {
    'Name': ['Alice', None, 'Bob'],
    'Age': [25, None, 30]
}

df = pd.DataFrame(data)

print(df.isnull())

print(df.fillna(0))

print(df.dropna())
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Sara'],
    'Gender': ['F', 'M', 'F'],
    'Age': [25, 30, 27]
}

df = pd.DataFrame(data)

print(df.groupby('Gender')['Age'].mean())
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'key': [1, 2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'key': [1, 2],
    'Age': [25, 30]
})

print(pd.merge(df1, df2, on='key'))
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
}

df = pd.DataFrame(data)

df.to_csv("output.csv", index=False)

new_df = pd.read_csv("output.csv")

print(new_df)
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Alice'],
    'Age': [25, 30, 25]
}

df = pd.DataFrame(data)

print(df.sort_values('Age'))

print(df['Age'].apply(lambda x: x + 1))

print(df['Name'].unique())

print(df['Age'].value_counts())
```
